```python
import matplotlib.pyplot as plt
```

```python
with open("mlfq_data_dump.txt", "r") as f:
    data = f.read().splitlines()
```

```python
ticks = [[] for i in range(100)]
q = [[] for i in range(100)]
for line in data:
    x = line.split()
    if (len(x) != 3):
        print(x)
        break
    ticks[int(x[0])].append(int(x[1]))
    q[int(x[0])].append(int(x[2]))
```

```python
legend = []
for i in range(100):
    if (len(ticks[i]) > 0):
        plt.plot(ticks[i], q[i])
        legend.append("pid "+str(i))
plt.xlabel("ticks")
plt.ylabel("queue")
plt.title("MLFQ Scheduling analysis (Aging threshold = 40)")
# changing size
fig = plt.gcf()
fig.set_size_inches(9, 6)
plt.legend(legend)
plt.show()
```
